```python
##lower(): Return A new string with all characters convert into lower case
##upper():Return A new string with all characters convert into upper case
```

```python
institute= "Edukron"
focus_area= "Data Engineer"
tech_stack= "Azure Data Engineer and Python"
```

```python
## Lower 
lower_institute= institute.lower()
print(lower_institute)

lower_focus_area= focus_area.lower()
print(lower_focus_area)

lower_tech_stack= tech_stack.lower()
print(lower_tech_stack)
```

```python
lower_institute= institute.lower()
print(f"Lower case Institute: '{lower_institute}")

lower_focus_area= focus_area.lower()
print(f"Lower case Focus Area: '{lower_focus_area}")

lower_tech_stack= tech_stack.lower()
print(f"Lower case Tech Stack: '{lower_tech_stack}")
```

```python
## Upper
upper_institute= institute.upper()
print(f"Upper case Institute: '{upper_institute}'")

upper_focus_area= focus_area.upper()
print(f"Upper case Focus Area: '{upper_focus_area}'")

upper_tech_stack= tech_stack.upper()
print(f"Upper case Tech Stack: '{upper_tech_stack}'")
```

```python
mixed_case = "PyThoN ProJEct"
mixed_lower = mixed_case.lower()
mixed_upper = mixed_case.upper()
print(f"Mixed Lower:'{mixed_lower}'")
print(f"Mixed Upper:'{mixed_upper}'")
```

```python
## numer and special characters are not affected by lower and upper
code="Batch-121-Data Engineer"
Lower_code= code.lower()
Upper_code= code.upper()
print(f"Upper Code:'{Upper_code}'")
print(f"Lower Code:'{Lower_code}'")
```

```python
##White Space= just unwanted space is called white space
##for remove the white space we use strip function
##strip([char]):= Returns new string with leading and trailing character removed
##lstrip([char]): removing the left spaces if the having
##rstrip([char]): removing the right spaces if they having
```

```python
user_input="    Data Engineer    "
cleaned_input = user_input.strip()
cleaned_input
```

```python
##lstrip
user_input="    Data Engineer    "
lstrip_user_input= user_input.lstrip()
lstrip_user_input
```
